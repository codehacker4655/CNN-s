```python
import tensorflow
from tensorflow.keras import layers,models
```

```python
x=layers.Input(shape=(224,224,3))
```

```python
conv11=layers.Conv2D(64,(3,3),activation='relu',padding='same')(x)
conv12=layers.Conv2D(64,(3,3),activation='relu',padding='same')(conv11)
pool1=layers.MaxPooling2D(pool_size=(2,2),padding='valid',strides=(2,2))(conv12)

conv21=layers.Conv2D(128,(3,3),activation='relu',padding='same')(pool1)
conv22=layers.Conv2D(128,(3,3),activation='relu',padding='same')(conv21)
pool2=layers.MaxPooling2D((2,2),padding='valid',strides=(2,2))(conv22)

conv31=layers.Conv2D(256,(3,3),activation='relu',padding='same')(pool2)
conv32=layers.Conv2D(256,(3,3),activation='relu',padding='same')(conv31)
conv33=layers.Conv2D(256,(3,3),activation='relu',padding='same')(conv32)
pool3=layers.MaxPooling2D((2,2),padding='valid',strides=(2,2))(conv33)

conv41=layers.Conv2D(512,(3,3),activation='relu',padding='same')(pool3)
conv42=layers.Conv2D(512,(3,3),activation='relu',padding='same')(conv41)
conv43=layers.Conv2D(512,(3,3),activation='relu',padding='same')(conv42)
pool4=layers.MaxPooling2D((2,2),padding='valid',strides=(2,2))(conv43)

conv51=layers.Conv2D(512,(3,3),activation='relu',padding='same')(pool4)
conv52=layers.Conv2D(512,(3,3),activation='relu',padding='same')(conv51)
conv53=layers.Conv2D(512,(3,3),activation='relu',padding='same')(conv52)
pool5=layers.MaxPooling2D((2,2),padding='valid',strides=(2,2))(conv53)

flat=layers.Flatten()(pool5)
dense1=layers.Dense(4096,activation='relu')(flat)
dense2=layers.Dense(4096,activation='relu')(dense1)
output=layers.Dense(1000,activation='softmax')(dense2)
```

```python
model=models.Model(inputs=x,outputs=output)
```
